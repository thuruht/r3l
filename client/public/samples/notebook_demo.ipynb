## Drift Analysis

Analyzing the signal network for resonance patterns.

```python
import numpy as np

def detect_resonance(signals, threshold=0.7):
    pairs = []
    for i, a in enumerate(signals):
        for b in signals[i+1:]:
            if abs(a['freq'] - b['freq']) < threshold:
                pairs.append((a['id'], b['id']))
    return pairs

print(f'Loaded {len(signals)} signals from drift network')
print(f'Resonance threshold: {threshold:.2f}')
```

```python
signals = [
    {'id': 'relay_42', 'freq': 0.74},
    {'id': 'signal_wolf', 'freq': 0.68},
    {'id': 'drift_runner', 'freq': 1.20},
    {'id': 'echo_9', 'freq': 1.15},
]

pairs = detect_resonance(signals)
pairs
```
